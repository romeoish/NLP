```python
df = df.rename(columns={'가입번호':'entr_no', '고객번호':'cust_no', '실시간TV채널명':'channel', '실시간TV프로그램시작일시':'show_date', '시청시간':'view_period', 
    '홈쇼핑제품명':'item_name', '브랜드명':'brand', '제품가격':'price'})
df['show_date'] = df.show_date.apply(lambda date: str(date)[:-4])
df.show_date = pd.to_datetime(df.show_date, format='%Y%m%d%H')
df.drop('entr_no', axis=1, inplace=True)
df
```

```python
measure = r'[0-9]+종세트|[0-9]+s\ss|[0-9]+ss|[0-9]+p|[0-9]+캡슐|[0-9]+환|[0-9]+인분|[0-9]+원|[0-9]+장|[0-9]+종|[0-9]+형|[0-9]+만원|[0-9]+ml|[0-9]+ea|[0-9]+set|[0-9]+매|[0-9]+개입|[0-9]+롤|[0-9]+박스입|[0-9]+박스분|[0-9]+박스|[0-9]+포|[0-9]+일분|[0-9]+주분|[0-9]+인용|[0-9]+채|[0-9]+겹|[0-9]+차|[0-9]+개월분|[0-9]+보틀|[0-9]+kg|[0-9]+g|[0-9]+cm|[0-9]+팩|[0-9]+l|[0-9]+리터|[0-9]+봉|[0-9]+통|[0-9]+세트|[0-9]+병|[0-9]+주|[0-9]+개|[0-9]+일|[0-9]+캔|[0-9]+박|[0-9]+pcs'



def clean_text(text):
    text = str(text)
    text = re.sub(r'[^ A-Za-z0-9가-힣+\[\]\(\)]', ' ', text)
    text = re.sub('\(+[가-힣0-9a-zA-Z\-_, ]+\)', '', text)
    text = re.sub('\[+[가-힣0-9a-zA-Z\-_, ]+\]', '', text)
    review = re.sub(r'[@%\\*=()/~#&\+á?\xc3\xa1\-\|\.\:\;\!\-\,\_\~\$\+\'\"]', ' ',text) #remove punctuation
    review = review.lower() #lower case
    review = re.sub(measure,'',review)
    review = re.sub(r'\s\d+\s','', review)# remove number
    review = re.sub(r'\s[A-Za-z]\s', '', review)
    review = re.sub(r'\s+', ' ', review) #remove extra space
    review = re.sub(r'<[^>]+>','',review) #remove Html tags
    review = re.sub(r'\[|\]', ' ', review)
    review = re.sub(r'총{1}', '', review)
    review = re.sub(r'\s+', ' ', review) #remove spaces
    review = re.sub(r"^\s+", '', review) #remove space from start
    review = re.sub(r'\s+$', '', review) #remove space from the end
    review = re.sub(r'\s[A-Za-z]', ' ', review)
    review = re.sub(r'\s\d', ' ', review)
    review = re.sub(r'x', '', review)
    
    return review


def clean_text2(text):
    text = str(text)
    text = re.sub(r'[^ A-Za-z0-9가-힣]', ' ', text)
    review = re.sub(r'[@%\\*=()/~#&\+á?\xc3\xa1\-\|\.\:\;\!\-\,\_\~\$\+\'\"]', ' ',text) #remove punctuation
    review = review.lower() #lower case
    review = re.sub(r'\s+', ' ', review) #remove extra space
    review = re.sub(r'\[|\]', ' ', review)
    review = re.sub(r'\s+', ' ', review) #remove spaces
    review = re.sub(r"^\s+", '', review) #remove space from start
    review = re.sub(r'\s+$', '', review) #remove space from the end
    
    return review


def for_emd(text):
    item_name = text['cleaned_item']
    brand = text['brand']
    if brand == 'nan':
        emd_text = item_name
        
    else:
        emd_text = brand + ' ' + item_name
        
    return emd_text
```

```python
df['cleaned_item'] = df.item_name.apply(clean_text)
df['brand'] = df.brand.apply(clean_text2)
df = df.drop_duplicates()
df['emd_text'] = df.apply(for_emd, axis=1)
df = df.sort_values(by='show_date')
```
